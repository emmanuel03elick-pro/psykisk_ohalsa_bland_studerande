```python
import pandas as pd
import sqlite3
import os 
df = pd.read_csv("student_depression_dataset.csv")
```

```python
df.head(10)
```

```python

# Rename ALL columns

df_clean = df.rename(columns={
    'Gender': 'gender',
    'Have you ever had suicidal thoughts ?': 'suicidal_thoughts',
    'Family History of Mental Illness': 'family_history',
    'Dietary Habits': 'dietary_habits',
    'Sleep Duration': 'sleep_duration',
    'Profession': 'profession',
    'Financial Stress': 'financial_stress',
    'Work/Study Hours': 'work_study_hours',
    'Academic Pressure': 'academic_pressure',
    'Job Satisfaction': 'job_satisfaction',
    'Study Satisfaction': 'study_satisfaction',
    'Age': 'age',
    'City': 'city',
    'Work Pressure': 'work_pressure',
    'CGPA': 'cgpa',
    'Degree': 'degree',
    'Depression': 'depression'
})

# Apply all mappings 

df_clean['gender'] = df_clean['gender'].map({
  'Male': 1, 'Female': 0
  })

df_clean['suicidal_thoughts'] = df_clean['suicidal_thoughts'].map({
  'Yes': 1, 'No': 0
  })

df_clean['family_history'] = df_clean['family_history'].map({
  'Yes': 1, 'No': 0
  })

df_clean['dietary_habits'] = df_clean['dietary_habits'].map({
  'Unhealthy': 0,
  'Moderate': 1, 
  'Healthy': 2
  })

df_clean['sleep_duration'] = df_clean['sleep_duration'].str.strip("'").map({
    'Less than 5 hours': 4,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9
})
df_clean['profession'] = df_clean['profession'].str.strip().str.strip("'")

df_clean['financial_stress'] = pd.to_numeric(df_clean['financial_stress'], errors='coerce')
```

```python
import sqlite3
from pathlib import Path

db_path = Path("../databases/sqlite-dbs/student_depression.db")

conn = sqlite3.connect(db_path)
df_clean.to_sql("student_depression", conn, if_exists="replace", index=False)
conn.close()

print(f"Databas skapad: {db_path.resolve()}")
```
